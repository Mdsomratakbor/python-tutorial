# COVID-19 Trends with NumPy

*NumPy-first, minimal pandas; pure Matplotlib for charts.*


## Overview
- Load COVID-19 dataset from **Our World in Data** (OWID) or fall back to a small synthetic sample if offline.
- NumPy-first calculations: rolling averages (via convolution), growth rates, per-capita normalization, and simple projections.
- Pure Matplotlib for plotting.
